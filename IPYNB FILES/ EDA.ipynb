```python
def plots(df,var,t):
    plt.figure(figsize=(13,5))
    plt.subplot(121)
    sns.distplot(df[var])
    plt.title('before' + str(t))
    plt.subplot(122)
    sns.distplot(t)
    plt.title('After' + str(t))
```

```python
plots(df,'RH',np.log(df['RH']))
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
sns.distplot(X['DMC'])
plt.title('day DMC')

plt.subplot(122)
stats.probplot(X['DMC'], dist="norm", plot=plt)
plt.title('DMC QQ Plot')

plt.show()
```

```python
def plots(df,var,t):
    plt.figure(figsize=(13,5))
    plt.subplot(121)
    sns.kdeplot(df[var])
    plt.title('before' + str(t))
    plt.subplot(122)
    p1 = t.fit_transform(df[[var]]).flatten()
    sns.kdeplot(p1)
    plt.title('After' + str(t))
```

```python
## Applying Yeo johnson as our datasets has some 0 values also
for col in ['day', 'month', 'RH', 'Ws', 'Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI']:
    plots(X,col,PowerTransformer())
    
```
